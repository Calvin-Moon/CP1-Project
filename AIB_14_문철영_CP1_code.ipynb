```python
import nltk
from konlpy.tag import Okt; t = Okt()
```

```python
contents_l = review_low
contents_l
```

```python
review_low = ''

for each_line in contents_l:
    review_low = review_low + each_line + '\n'

review_low
```

```python
tokens_l = t.morphs(review_low)
tokens_l
```

```python
ko_low = nltk.Text(tokens_l)
ko_low.tokens
```

```python
len(set(ko_low.tokens))
```

```python
ko_low.vocab().most_common(100)
```

```python
stop_words = ['\n','.','이','에','가','도', '은','을', '잘','그냥','안','좋아요','생각','책상','으로','하고',',', '..', '요','이', '가','도','으로','을', '에','는','좀','다','보다','한','로',
               '했는데','를', '...', '상', '것','!','서', '많이','에서','거',  '합니다','해서','같아요','\n\n','조금','더','못','들','~',
               '비','입니다','때','고','제','만','?','하기', '하는','딱',  '이라', '할',  '그','과', '의','중','좋은', '씁니다', '다시',
               '했습니다','저',  '라','수','하는데','개', '인데','있어서', '좋습니다','귀찮아서', '만족합니다','면','내','하나', '두', '좋은데',
              '있습니다','적','걸','그래도','않아요','인','다만','왔어요','분','정말','나','깔끔하고','곳','에는','지', 'ㅠ','다른','하지만','전',
              '듯', '게','뭐','없이','엔','보니','2', '같은데','튼튼하고','괜찮아요','같습니다','네','....','함', '살짝','이나','또','번','있고','와서']
```

```python
tokens_l = [each_word for each_word in tokens_l
            if each_word not in stop_words]

ko_low = nltk.Text(tokens_l)
ko_low.vocab().most_common(100)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import graphviz
from sklearn.tree import export_graphviz

plt.rc('font', family='NanumBarunGothic', size= 15)
plt.figure(figsize=(25,10))
ko_low.plot(50)
plt.show()
```

```python
from collections import Counter
from wordcloud import WordCloud 

count = Counter(ko_low)
words = dict(count.most_common())

wordcloud = WordCloud(font_path="/Library/Fonts/NanumGothic.ttf",
                      relative_scaling = 0.2,
                      background_color='white').generate_from_frequencies(words)
plt.figure(figsize=(25,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
df_high = pd.concat([df_4, df_5], axis=0)
df_high
```

```python
review_high = df_high.content
contents_h = review_high

review_high = ''

for each_line in contents_h:
    review_high = review_high + each_line + '\n'

review_high
```

```python
tokens_h = t.morphs(review_high)

tokens_h = [each_word for each_word in tokens_h
            if each_word not in stop_words]

ko_high = nltk.Text(tokens_h)

plt.rc('font', family='NanumBarunGothic', size= 15)
plt.figure(figsize=(25,10))
ko_high.plot(50)
plt.show()
```

```python
count = Counter(ko_high)
words_h = dict(count.most_common())

wordcloud_h = WordCloud(font_path="/Library/Fonts/NanumGothic.ttf",
                      relative_scaling = 0.2,
                      background_color='white').generate_from_frequencies(words_h)
plt.figure(figsize=(25,10))
plt.imshow(wordcloud_h)
plt.axis("off")
plt.show()
```
